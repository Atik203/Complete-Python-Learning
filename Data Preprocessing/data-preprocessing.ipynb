## Data Preprocessing
### Step of preprocessing data before analysis or modeling.
1. **Import Libraries**: Load necessary libraries for data manipulation and analysis.
2. **Read Dataset**: Load the dataset into a DataFrame.
3. **Sanity Check**: Inspect the dataset for missing values, duplicates, and data types.
4. **Exploratory Data Analysis (EDA)**: Perform initial analysis to understand data distribution and relationships.
5. **Missing Value Treatment**: Handle missing values through imputation or removal.
6. **Outlier Detection and Treatment**: Identify and manage outliers in the dataset.
7. **Duplicate Removal**: Remove duplicate records to ensure data quality.
8. **Normalization/Standardization**: Scale features to a standard range or distribution.
9. **Encoding of Data**: Convert categorical variables into numerical format using techniques like one-hot encoding or label encoding.

### 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Read Dataset

```python
df = pd.read_csv("Life Expectancy Data.csv")
```

```python
df.head()
```

```python
df.tail()
```

### 3. Sanity Check

```python
df.shape
```

```python
df.info()
```

```python
# finding missing values
df.isnull().sum()
```

```python
# finding the percentage of missing values
(df.isnull().sum()/df.shape[0])*100
```

```python
## finding duplicate values
df.duplicated().sum()
```

```python
## finding garbage values
for column in df.select_dtypes(include=['object']).columns:
    print(df[column].value_counts())
    
```

```python
### 4.Exploratory Data Analysis 
```

```python
# descriptive statistics
df.describe()
```

```python
## descriptive statistics for object type columns
df.describe(include=['object'])
```

```python
## histogram for understanding distribution of numerical columns
for column in df.select_dtypes(include=['int64','float64']).columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df[column],kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
## Box plot for identifying outliers
for column in df.select_dtypes(include=['int64','float64']).columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(y=df[column])
    plt.title(f'Box plot of {column}')
    plt.show()
```
